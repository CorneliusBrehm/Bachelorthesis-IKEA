# Load and View Tables

## Store Info

```python
# Establish connection to the database
conn = sqlite3.connect(dbPath)
cursor = conn.cursor()

# Select all rows from the table
cursor.execute('SELECT * FROM storeInfo')

# Fetch all rows and print the results
rows = cursor.fetchall()

#---

# Retrieve column names of the table
cursor.execute("PRAGMA table_info('storeInfo')")
columns = cursor.fetchall()

# Extract column names from the result
columnNames = [column[1] for column in columns]

dfStoreInfos = pd.DataFrame(rows, columns = columnNames)

#---

# Close the connection
conn.close()

dfStoreInfos
```
